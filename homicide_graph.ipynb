```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

outputDir = "./output"
```

### 1.  States with higher number of unsolved crimes

```python
unsolvedCrimesPerState = pd.read_csv(f"{outputDir}/1_unsolvedCrimesPerState.csv", header=0)
```

```python
fig, ax = plt.subplots()
 
ax.bar(unsolvedCrimesPerState.State, unsolvedCrimesPerState["count"], label='Homicides')
ax.legend()
fig.suptitle('Unsolved crimes')
plt.show()
```

### 2. Most violent years of the 21st century

```python
mostViolentYear21thCentury = pd.read_csv(f"{outputDir}/2_mostViolentYear21thCentury.csv", header=0)
```

```python
fig, ax = plt.subplots()

ax.plot(mostViolentYear21thCentury.Year, mostViolentYear21thCentury.Count, 'r-', label='Homicides')
ax.legend()
fig.suptitle('Most violent year')
plt.show()
```

### 3. Number of crimes per gender

```python
crimesPerGender = pd.read_csv(f"{outputDir}/3_crimesPerGender.csv", header=0)
```

```python
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(crimesPerGender.Count, labels=crimesPerGender.VictimSex,
                                  shadow=True, startangle=90, autopct='%1.1f%%',
                                  explode=(0, 0.1, 0.1))

ax.legend(wedges, crimesPerGender.VictimSex,
          title="Genders",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

fig.suptitle('Crimes per gender')
plt.show()
```

### 8. Unsolved crimes by race

```python
unsolvedByRace = pd.read_csv(f"{outputDir}/8_unsolvedByRace.csv", header=0)
```

```python
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(unsolvedByRace["count"], labels=unsolvedByRace.victimRace,
                                  shadow=True, startangle=90, autopct='%1.1f%%', pctdistance=0.85,
                                  explode=(0.05,0.05,0.05,0.05,0.05))

ax.legend(wedges, unsolvedByRace.victimRace,
          title="Races",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

for text in texts:
    text.set_visible(False)
    
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

fig.suptitle('Unsolved crimes by race')
plt.show()
```

### 9. Homicides by perpetrator age range

```python
crimesByPerpetratorAgeRange = pd.read_csv(f"{outputDir}/9_crimesByPerpetratorAgeRange.csv", header=0)
crimesByPerpetratorAgeRange = crimesByPerpetratorAgeRange.sort_values('PerpetratorAgeRange')
```

```python
fig, ax = plt.subplots()
ax.bar(crimesByPerpetratorAgeRange.PerpetratorAgeRange, crimesByPerpetratorAgeRange.Crimes, label='Homicides', color='b')
ax.legend()
fig.suptitle('Homicides by perpetrator age range')
plt.xticks(rotation=45, ha='right');
plt.show()
```
